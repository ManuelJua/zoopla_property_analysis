```python
import pandas as pd
import requests
from urllib.parse import urlencode
import numpy as np

#Here goes the file address with info for geocoding
geolocation_file_address=''
df=pd.read_csv(geolocation_file_address)
```

```python
with open('geolocation_token.txt','r') as file:
    API_KEY=file.read()
format='json'

for i in df[df['latitude'].isna()].index:
    address=df['address'].iloc[i]
    params={'address':address,'key':API_KEY}
    url_params=urlencode(params)
    url_base=f'https://maps.googleapis.com/maps/api/geocode/{format}?{url_params}'

    try:
        response=requests.get(url_base)
        df.loc[i,'latitude']=response.json()['results'][0]['geometry']['location']['lat']
        df.loc[i,'longitude']=response.json()['results'][0]['geometry']['location']['lng']
        print('{} --- {} --- lat: {} --- lng: {}'.format(i,df.loc[i,'address'],df.loc[i,'latitude'],df.loc[i,'longitude']))
        
    except:
        df.loc[i,'latitude']=None
        df.loc[i,'longitude']=None

```

```python
df.to_csv('zoopla_edinburgh_with_geoloc.csv',index=False)
```
